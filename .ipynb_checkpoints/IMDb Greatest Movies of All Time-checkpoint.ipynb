```python
import requests
```

<b>Step 2:- Get the content of the webpage using requests library</b>

```python
movie_url = 'https://www.imdb.com/list/ls055592025/'
```

```python
resObj = requests.get(movie_url)
```

```python
resObj.status_code #200 means that the request was successful
```

<h5>Grabbing the response obtained in the above step and writing it to a webpage html file.</h5>

```python
snip = resObj.text
```

```python
with open('web.html','w',encoding='utf-8') as fs:
    fs.write(snip)
```

<b>Step 3:- Install BeautifulSoup for parsing information about various components of the webpage</b>

```python
!pip install beautifulsoup4 --upgrade --quiet
```

```python
from bs4 import BeautifulSoup
```

<b>Step 4:- Fetch the contents of the webpage using BeautifulSoup class</b>

```python
webDoc = BeautifulSoup(snip,'html.parser')
```

```python
type(webDoc)
```

<b>Grabbing the header tag of each movie</b>

```python
headerClass = 'lister-item-header'
tmp1 = webDoc.find_all('h3',{'class':headerClass})
```

```python
headerTags = [i.find('a') for i in tmp1 if i is not None]
headerTags
headerTags2 = [i.text for i in headerTags]
headerTags2[:10]
```

<b>Grabbing the description data of each movie</b>

```python
divContentList = webDoc.find_all('div',{'class':'lister-item-content'})
```

```python
desc_tags = []
for i in divContentList:
    for j in i.find_all('p',{'class':''}):
        desc_tags.append(j.text.strip())
```

```python
desc_tags[:2] # description data of first two movies
```

```python
len(desc_tags) #as we are dealing with 100 movies, so we get 100 desc_tags
```

<b>Grabbing the webpage link of each movie.</b>

```python
moviePageLinks = []
baseUrl = 'https://imdb.com'
for i in headerTags:
    moviePageLinks.append(baseUrl + i['href'])
```

<b>Links of the top 10 movies</b>

```python
top10 = moviePageLinks[:10]
for i in top10:
    print(i)
```

<b>Grabbing the release year</b>

```python
releaseYear = webDoc.find_all('span',{'class':'lister-item-year text-muted unbold'})
releaseYear = [i.text[1:-1] for i in releaseYear]
```

**Grabbing the rating score**

```python
ratingDiv = webDoc.find_all('div',{'class':'ipl-rating-star small'})
ratingScoreList = []
for i in ratingDiv:
    ratingScoreList.append(i.find('span',{'class':'ipl-rating-star__rating'}))
ratingScoreList = [i.text for i in ratingScoreList]
ratingScoreList[:10]
```

<b>Grabbing the director's name</b>

```python
temp = []
for i in divContentList:
    temp.extend(i.find_all('p',{'class':'text-muted text-small'}))
directorList = []
for i in temp:
    directorList.append(i.find('a'))
directorList=[i.text for i in directorList if i is not None]
directorList[:10]
```

<b>Grabbing the Metascore</b>

```python
divContentList
metaScore = []
for i in divContentList:
    if len(i.find_all('div',{'class':'inline-block ratings-metascore'})):
        metaScore.append(i.find('div',{'class':'inline-block ratings-metascore'}).find('span',{'class':'metascore favorable'}).text.strip())
    else:
        metaScore.append('N/A')
metaScore[:10]
```

<b>Step 5:- Install Pandas for creating a dataframe of the extracted data</b>

```python
!pip install pandas --upgrade --quiet
```

```python
import pandas as pd
```

<b>Step 6:- Create a Dataframe from the extracted data in the form of lists</b>

```python
datasetDict = {'Title':headerTags2,'Year':releaseYear,'Score':ratingScoreList,'Meta Score': metaScore,'Director':directorList, 'Movie Link': moviePageLinks}
```

```python
df = pd.DataFrame(datasetDict)
df.index=[str(i) for i in range(1,101)]
```
